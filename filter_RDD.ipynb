```python
from pyspark import SparkContext
from pyspark.sql import SQLContext
```

```python
sc = SparkContext(master = "local", appName="FilteringBoyaca")
```

```python
sqlContext = SQLContext(sc)
```

```python
!ls
```

```python
reg2 = sc.textFile("TOTAL_NACIONAL_REG2.txt",2).map(lambda line : line.split("\t"))
```

```python
reg2.countApprox(20)
```

```python
type(reg2)
```

```python
reg2.take(5)
```

```python
reg_boy = reg2.filter(lambda x: x[0] == '15')
```

```python
reg_boy.countApprox(20)
```

```python
type(reg_boy)
```

```python
reg_boy.take(5)
```

```python
cols = reg2.take(1)[0]
cols
```

```python
df = sqlContext.createDataFrame(reg_boy, cols)
```

```python
df.show()
```

```python
# Write CSV 
df.collect().write.format('com.databricks.spark.csv').options(header='true').save('BASE_BOYACA.csv')
```
